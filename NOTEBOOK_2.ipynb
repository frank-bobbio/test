```python
"""
print_dimesiones_del_df()
Recibe un dataframe e imprime el número de filas y columnas del mismo.
"""

def print_dimesiones_del_df(df):
  print(f'Dimensiones del dataframe individual: Numero de filas: {len(df.index)}, Número de columnas {len(df.columns)}')

"""
print_dimesiones_datasets()
Recibe un diccionario con varias dataframes e imprime el número de filas y columnas
de cada dataframe en el diccionario.
"""

def print_dimesiones_datasets(df_dict):
  print("-----------Datasets en el diccionario---------")
  for key, df in df_dict.items():
    print(f"{key}: Numero de filas: {len(df.index)}, Número de columnas {len(df.columns)}")
  print("----------------------------------------------")
  #return pd.Series([len(df.index) for df in df_dict.values()])

"""
elimino_col_parcialmente_nulas()
Elimina columnas parcialmente vacias en un dataframe.
El porcentaje que utiliza como umbral para eliminar las columnas lo recibe como parámetro.
"""

def elimino_col_parcialmente_nulas(df, percent):
  columnas_vacias_parcialmente = [col for col in df.columns if (df[col].isnull().sum()*100/len(df)>percent)]
  #print(columnas_vacias_parcialmente)
  #print(len(columnas_vacias_parcialmente))
  return df.drop(columns=columnas_vacias_parcialmente)

def elim_filas_repes(df):
  if df.shape[0] - df.drop_duplicates().shape[0] > 0:
      print(f"Number of duplicate rows deleted: {df.shape[0] - df.drop_duplicates().shape[0]}")
  else: print("There is not a duplicated rows")

def delete_unique_col(df):
  for col in df.columns:
      if len(df[col].unique()) == 1:
        df = df.drop(col, axis=1)
  print(f"Shape after try to eliminate columns with one unique value: {df.shape}")


"""
eliminar_no_numericos(df, col)

Función para eliminar los datos no numéricos (Ejemplo /N) en una columna de un dataframe.
Devuelve un nuevo dataframe con las filas donde la columna tiene valores numéricos.
"""
def eliminar_no_numericos(df, col):

  # Se eliminan los valores NaN
  mask = df[col].notna()
  df_temp = df[mask]

  # Es necesario asegurarse que el tipo de la columna sea string sino puede devolver NaN en lugar de
  # true o false como se espera en función de que sea numérico o no.
  mask = df_temp[col].astype(str).str.isnumeric()
  df_temp = df_temp[mask]
  return df_temp


```

## Funciones

A continuación, se definen una serie de funciones que se reutilizarán para la carga de los datos asociados a las diferentes tablas.

Lista de funciones a definir:
*   cargar_systrack_csv
*   proc_systrack
*   cargar_etiquetas_csv
*   unifico_dataframes

```python

"""
Carga los dataframes de systrack desde los archivos CSV.
Recibe como parámetros la ubicación de los archivos (PATH) y el formato del nombre de las tablas.
Se debe ejecutar la función para cada tipo de tabla (RTP_BIOS, RTP_BATTERY, etc).
Ejemplo de tabla "*RPT_BIOS*.csv"
"""
def cargar_systrack_csv(PATH, tabla, solo_CXB, d, filtro, numericos, mask='', dtype={}):

  csv_files = glob.glob(os.path.join(PATH, tabla))

  for f in csv_files:
    key = f.split("/")[-1]

    # Si existen columnas de tiempo se importan con el tipo de datos correcto.
    # Las 2 columnas de tiempo que aparecen juntas son DOTY y VWTIME
    if 'DOTY' in filtro:
      df_temp = pd.read_csv(f, sep=';', usecols=filtro, parse_dates=['DOTY', 'VWTIME'], dtype=dtype)
    else:
      df_temp = pd.read_csv(f, sep=';', usecols=filtro)
    if not (mask==''):
      #Pendiente a futuro. Cambiar condición por mask
      df_temp = df_temp[(df_temp['WTYPE_ID']==0)]

    # Los campos numéricos se pasan a tipo float y cuando encuentra un valor que no puede
    # convertir se pasa a NaN (a veces aparecen valores //N en campos numericos de las tablas)
    df_temp[numericos] =df_temp[numericos].apply(pd.to_numeric, errors='coerce')

    # Float64 da errores en pycaret
    df_temp[numericos] =df_temp[numericos].astype('float32')
    d[key] = proc_systrack(df_temp, solo_CXB=solo_CXB, filtro=filtro)



"""
  proc_systrack()
  Función para procesar los datafrmaes que provienen de systrack
  A partir de un dataframe de Systrack que contiene una colomna SystemId se encuentra el NM (número de máquina).
  Este número es el que nos permitirá vincular al info de Systrack con la de sustituciones del
  proveedor de mantenimiento.
  Ejemplo para el SystemID CXB0920346.LACAIXA.GLC.ES	tenemos que en número de máquina (NM) es 0920346
  Opcionalmente se pueden filtrar las maquinas de Caixa
  """
def proc_systrack(df_systrack, solo_CXB, filtro):

  df_systrack = df_systrack[filtro]

  # Se eliminan las columnas que tienen todos valores null
  df_systrack = df_systrack.dropna(axis=1, how='all')

  # Se eliminan las columnas con valores constantes
  df_systrack = df_systrack.loc[:,df_systrack.nunique() != 1]

  # Se filtran únicamente los SystemId que corresponden ca Caixa
  if solo_CXB:
    df_systrack = df_systrack.dropna(subset = ['SystemId'])
    mask = df_systrack['SystemId'].str.contains("CXB0")
    df_systrack = df_systrack[mask]

  return df_systrack


"""
cargar_etiquetas_csv()
Carga los dataframes con el resumen de las sustituciones desde los archivos CSV.

"""
def cargar_etiquetas_csv(PATH, tabla, d):

  csv_files = glob.glob(os.path.join(PATH, tabla))

  for f in csv_files:
    key = f.split("/")[-1]
    d[key] = pd.read_csv(f, sep=';', encoding='latin-1')
```

```python
"""
Función para diagnóstico que devuelve los valores infinitos en un dataframe.
"""
def num_infinitos(df):
  print("Número de inifnitos en el df:\n")
  return df[df == np.inf].count()
```

# Generación del dataframe Unificado de entrenamiento.

En esta instancia genera el dataframe unificado.

Se deberá recorrer todas las tablas seleccionadas con sus medidas para procesarlas y generar las características adecuadas. Adicionalmente se procesará el archivo que contiene los reemplazos de las máquinas para poder agregar las etiquetas.

## Funciones

A continuación, se definen una serie de funciones que se reutilizarán para la generación de las características asociadas a las diferentes tablas.

Lista de funciones:

*   unifico_dataframes
*   map_wguid_to_SystemId
*   agregar_NM
*   dataset_mas_reciente
*   ultimos_n_dias
*   agregado_caracteristicas
*   agregado_caracteristicas_2







```python
"""
unifico_dataframes()

A partir de un diccionario de dataframes de la misma tabla (ejemplo RAW_dfs_RPT_SV_HARDWARE)
se genera uno solo dataframe que agrupa todos los registros incluidos en los diferentes dataframes
del diccionario.
Esto permite tener todas las extracciones de una misma tabla en un solo
dataframe lo que simplifica el tratamiento posterior de la información.

"""

def unifico_dataframes(df_dict, elimino_duplicados=True):
  primero = True
  for value in df_dict.values():
    print_dimesiones_del_df(value)
    if primero:
        df_unificado = value
        primero = False
    else:
        if elimino_duplicados:
          # Se agrega solamente las maquinas que nuevas (o que no estan en el df_unificado)
          value = value[~value.index.isin(df_unificado.index)]
          df_unificado.append(value)
        else:
          df_unificado = df_unificado.append(value)

  print(f"Dimensiones del dataframe unificado:", end=" ")
  print(f"Numero de filas: {len(df_unificado.index)}, Número de columnas {len(df_unificado.columns)}")

  return df_unificado


"""
map_wguid_to_SystemId()
Permite obtener un mapeo entre los campos WGUID y SystemId.
Algunas tablas nos contienen el campo SystemId fundamente para determinar el número de máquina.
Para los casos anteriores se obtendrá el SystemId usando este dataframe generado que contiene el
mapeo entre WGUID y SystemID.
"""
def map_wguid_to_SystemId(df):

  df_wguid_SystemId = df[['WGUID', 'SystemId']]

  # Algunos dataset cambiaron el formato del WGUI poniendolo entre llaves por lo que se deben eliminar
  df_wguid_SystemId['WGUID'] = df['WGUID'].str.replace("[{}]","")

  # Se eliminan los duplicados
  df_wguid_SystemId = df_wguid_SystemId.drop_duplicates()
  print_dimesiones_del_df(df_wguid_SystemId)

  #df_wguid_SystemId = df_wguid_SystemId.set_index('WGUID')

  return df_wguid_SystemId


"""
agregar_NM()
Se agrega el número de máquina NM a un dataframe que no tenga esta columna.
Se puede hacer usando el SystemId. Si no está presente esta columna se genera a partir del WGUID.
"""
def agregar_NM(df, drop_NM_dup=False ,df_map_wguid_to_SystemId=False):

  # Algunos dataset cambiaron el formato del WGUI poniendolo entre llaves por lo que se deben eliminar
  df['WGUID'] = df['WGUID'].str.replace("[{}]","")

  if 'SystemId' not in df:
    df = pd.merge(df, df_map_wguid_to_SystemId, on='WGUID', how='left')

  # Agrego la columna NM obteniendo a partir de la info de SystemID
  df_temp = df.filter(['SystemId'])
  df['NM'] = df_temp['SystemId'].str.extract('(\d+)')

  # Se eliminan las filas que tienen NM nulo. Drop rows that has NaN values on selected columns
  df = df.dropna(subset = ['NM'])

  # La columna NM se pasa a tipo entero, es importante para eliminar el primer 0 (ya que servicenow no lo tiene)
  df = df.astype({'NM':'uint32'})

  # Opción de elimiar NM duplicados. Evaluar dependiendo del dataset si se debe hacer o no.
  if drop_NM_dup:
    df = df.drop_duplicates(subset='NM')

  # Seteo la columna NM como índice
  df = df.set_index('NM')

  return df

"""
dataset_mas_reciente()
Función para determinar cuál es el dataset más reciente de los que están en un diccionario.
En un diccionario se pueden almacenar varios datasets. La key será el nombre que contiene la fecha
en la que se tomaron los datos del dataset.
Se extrae la fecha del nombre en cada key y se encuentra la mas reciente.
"""
def dataset_mas_reciente(df_dict):
  date_time_str = '00/10/30'
  date_time_obj = datetime.strptime(date_time_str, '%y/%m/%d').date()
  fecha_ultimos_datos = date_time_obj

  for key in df_dict.keys():
    #print(key)
    match = re.search(r'\d{4}_\d{2}_\d{2}', key)
    print(match.group(0))
    date = datetime.strptime(match.group(), '%Y_%m_%d').date()
    if date > fecha_ultimos_datos:
      fecha_ultimos_datos = date
      key_mas_reciente = key

  print(f"Dataset más reciente {key_mas_reciente}")
  return key_mas_reciente
```

```python
"""
ultimos_n_dias()
Recibe un df y el número de días. El df es una tabla de systrack con el campo DOTY
donde almacena la fecha de las medidas (ej, CPU, MEM, etc) para cada máquina. La función filtra
las medidas de los últimos n días para cada maquina y devuelve ese df filtrado.
"""

def ultimos_n_dias(df, n, columna_tiempo):
  # Se genera ua serie donde para cada NM se obtiene la fecha máxima. (DOTY)
  ult_fecha = df.groupby('NM')[columna_tiempo].transform(max)

  # Se genera un array boleano donde para cada NM de verdadero o falso dependiendo si la fecha DOTY
  # es mayor la la última fecha menos n dias.
  mask_una_semana = df[columna_tiempo] >= ult_fecha - timedelta(days=n)

  # Se devuelve el df filtrado donde se tienen las medidas de los últimos n días para cada NM.
  return df[mask_una_semana]


"""
agregado_caracteristicas()
Permite agregar la característica pasada como parámetro al df_unificado, ésto utilizando
distinas ventanas de tiempo 1 semana (1W), 2 semanas (2W) y 1 mes (1M).
"""
def agregado_caracteristicas(df_origen, caract, columna_tiempo, relleno_nans = True):

  # Se agrupa por número de máquina para realizar las agregaciones para cada una de ellas

  deb = False

  global df_unificado

  # ---------------------AGREGACIÓN VENTANA DE 1 SEMANA---------------------------

  # Se filtran los últimos 7 días para cada maquina
  #df = top_n(df_origen, n=7, col='DOTY')

  df = ultimos_n_dias(df_origen, 7, columna_tiempo)

  # Se agrupa por NM y ordena por fecha.

  agrupados_por_NM = df[[columna_tiempo, caract]].groupby('NM')

  df = agrupados_por_NM.min()
  df = df.rename(columns={caract:caract + '_1W_min', columna_tiempo: columna_tiempo +'_1W_' + caract})
  df[caract + '_1W_mean'] = agrupados_por_NM.mean()[caract]
  df[caract + '_1W_std'] = agrupados_por_NM.std()[caract]
  df[caract + '_1W_max'] = agrupados_por_NM.max()[caract]
  df_unificado = pd.merge(df_unificado, df, left_index=True, right_index=True, how='left')

  if relleno_nans:
    if deb: print('relleno 1W')
    print('relleno 1W')
    df_unificado.fillna({caract + '_1W_min': 0, caract + '_1W_mean': 0,
            caract + '_1W_std': 0, caract + '_1W_max': 0}, inplace=True)

  # ------------------ FIN AGREGACIÓN VENTANA DE 1 SEMANA ---------------------------


  # ------------------ AGREGACIÓN VENTANA DE 2 SEMANAS ---------------------------

  # Se filtran los últimos 14 días de cada
  #df = top_n(df_origen, n=14, col='DOTY')
  df = ultimos_n_dias(df_origen, 14, columna_tiempo)


  agrupados_por_NM = df[[columna_tiempo, caract]].groupby('NM')
  df = agrupados_por_NM.min()
  df = df.rename(columns={caract: caract + '_2W_min', columna_tiempo: columna_tiempo + '_2W_' + caract})
  df[caract + '_2W_mean'] = agrupados_por_NM.mean()[caract]
  df[caract + '_2W_std'] = agrupados_por_NM.std()[caract]
  df[caract + '_2W_max'] = agrupados_por_NM.max()[caract]
  df_unificado = pd.merge(df_unificado, df, left_index=True, right_index=True, how='left')

  if relleno_nans:
    if deb: print('relleno 2W')
    print('relleno 2W')
    df_unificado[caract + '_2W_min'].fillna(df_unificado[caract + '_1W_min'], inplace=True)
    df_unificado[caract + '_2W_mean'].fillna(df_unificado[caract + '_1W_mean'], inplace=True)
    df_unificado[caract + '_2W_std'].fillna(df_unificado[caract + '_1W_std'], inplace=True)
    df_unificado[caract + '_2W_max'].fillna(df_unificado[caract + '_1W_max'], inplace=True)

  # ------------------ FIN AGREGACIÓN VENTANA DE 2 SEMANAS ---------------------------


  # --------------------- AGREGACIÓN VENTANA DE 1 MES ---------------------------

  # Se filtran los últimos 30 días de cada
  #df = top_n(df_origen, n=30, col='DOTY')
  df = ultimos_n_dias(df_origen, 30, columna_tiempo)


  # Para poder agregar el último valor medido de la característica, previo a agrupar se ordena por fecha.
  agrupados_por_NM = df[[columna_tiempo, caract]].groupby('NM')
  df = agrupados_por_NM.min()
  df = df.rename(columns={caract: caract + '_1M_min', columna_tiempo: columna_tiempo + '_1M_' + caract})
  df[caract + '_1M_mean'] = agrupados_por_NM.mean()[caract]
  df[caract + '_1M_std'] = agrupados_por_NM.std()[caract]
  df[caract + '_1M_max'] = agrupados_por_NM.max()[caract]
  df_unificado = pd.merge(df_unificado, df, left_index=True, right_index=True, how='left')

  if relleno_nans:
    if deb: print('relleno 1M')
    print('relleno 1M')
    df_unificado[caract + '_1M_min'].fillna(df_unificado[caract + '_2W_min'], inplace=True)
    df_unificado[caract + '_1M_mean'].fillna(df_unificado[caract + '_2W_mean'], inplace=True)
    df_unificado[caract + '_1M_std'].fillna(df_unificado[caract + '_2W_std'], inplace=True)
    df_unificado[caract + '_1M_max'].fillna(df_unificado[caract + '_2W_max'], inplace=True)


  # ------------------ FIN AGREGACIÓN VENTANA DE 1 MES ---------------------------

  # Agregado con el total de las medidas
  df = df_origen

  # Para poder agregar el último valor medido de la característica, previo a agrupar se ordena por fecha.
  agrupados_por_NM = df[[columna_tiempo, caract]].sort_values(by=columna_tiempo).groupby('NM')
  df = agrupados_por_NM.min()
  df = df.rename(columns={caract: caract + '_total_min', columna_tiempo: columna_tiempo + '_total_' + caract})
  df[caract + '_total_mean'] = agrupados_por_NM.mean()[caract]
  df[caract + '_total_std'] = agrupados_por_NM.std()[caract]
  df[caract + '_total_max'] = agrupados_por_NM.max()[caract]

  # Se agrega el último valor medido de la característica, este valor se calcula en la ventana mas extensa de tiempo.
  df[caract + '_ult_registro'] = agrupados_por_NM.nth(-1)[caract]
  df[caract + '_fecha_registro_mas_reciente'] = agrupados_por_NM.nth(-1)[columna_tiempo]
  df_unificado = pd.merge(df_unificado, df, left_index=True, right_index=True, how='left')

```

```python
"""
Función para la agregación de las características de la tabla
RPT_VUAPPFAULTSDAILY2 (FAULTS, HANGS y BSODS).
Estas características tienen una agregación distinta a las anteriores por lo que se
utiliza una función particular para realizar esta tarea.

"""
def agregado_caracteristicas_2(df_origen, caract_d, columna_tiempo):

  # Se agrupa por número de máquina para realizar las agregaciones para cada una de ellas

  # Se cuentan los flags y se suman los NUMFLTS

  #deb = False
  deb = True
  relleno_nans = True

  global df_unificado

  if caract_d == 'FAULTS':
    flag_val = 1
  if caract_d == 'HANGS':
    flag_val = 2
  if caract_d == 'BSODS':
    flag_val = 3

  # ---------------------AGREGACIÓN VENTANA DE 1 SEMANA---------------------------

  # Se filtran los últimos 7 días de cada
  #df = top_n(df_origen, n=7, col='DOTY')
  df = ultimos_n_dias(df_origen, 7, columna_tiempo)


  mask_flag = df['FLAGS']==flag_val
  #mask = mask_1W & mask_flag
  mask = mask_flag

  #Se dropean los duplicados (teniendo en cuenta la appname y se agrupan por NM)
  agrupados_por_NM = df[mask].drop_duplicates().groupby('NM')
  df_1 = agrupados_por_NM.agg({'FLAGS': 'count', 'NUMFLTS': 'sum'})
  if deb: print('filtro_1')
  #if deb: print(df_1)

  agrupados_por_NM = df[~mask].groupby('NM')
  f = lambda x: 0
  df_2 = agrupados_por_NM.agg({'FLAGS': f, 'NUMFLTS': f})
  if deb: print('filtro_2')
  #if deb: print(df_2)

  df_2.update(df_1)
  df_2 = df_2.rename(columns = {'FLAGS': caract_d + '_1W', 'NUMFLTS': 'NUMFLTS_1W_' + caract_d})

  # if deb: print(df_2)

  if relleno_nans:
      if deb: print('relleno 1W')
      #df_2[caract_d + '_1W'].fillna(0, inplace=True)
      df_2.fillna({caract_d + '_1W': 0, 'NUMFLTS_1W_' + caract_d: 0}, inplace=True)


  df_unificado = pd.merge(df_unificado, df_2, left_index=True, right_index=True, how='left')
  if deb: print('fin 1W')

  # ------------------ FIN AGREGACIÓN VENTANA DE 1 SEMANA ---------------------------


  # ------------------ AGREGACIÓN VENTANA DE 2 SEMANAS ---------------------------

  # Se filtran los últimos 14 días de cada
  df = ultimos_n_dias(df_origen, 14, columna_tiempo)

  mask_flag = df['FLAGS']==flag_val
  mask = mask_flag

  #mask = mask_2W & mask_flag
  #agrupados_por_NM = df_origen[mask].groupby('NM')
  #Se dropean los duplicados (teniendo en cuenta la appname y se agrupan por NM)
  agrupados_por_NM = df[mask].drop_duplicates().groupby('NM')
  df_1 = agrupados_por_NM.agg({'FLAGS': 'count', 'NUMFLTS': 'sum'})
  if deb:  print('filtro_2W_1')

  agrupados_por_NM = df[~mask].groupby('NM')
  f = lambda x: 0
  df_2 = agrupados_por_NM.agg({'FLAGS': f, 'NUMFLTS': f})
  if deb:  print('filtro_2W_2')

  df_2.update(df_1)
  df_2 = df_2.rename(columns = {'FLAGS': caract_d + '_2W', 'NUMFLTS': 'NUMFLTS_2W_' + caract_d})

  if relleno_nans:
    df_2.fillna({caract_d + '_2W': 0, 'NUMFLTS_2W_' + caract_d: 0}, inplace=True)

  # previo al merge relleno todos los NaNs con ceros
  df_2 = df_2.fillna(0)

  df_unificado = pd.merge(df_unificado, df_2, left_index=True, right_index=True, how='left')
  if deb: print('fin 2W')

  # ------------------ FIN AGREGACIÓN VENTANA DE 2 SEMANAS --------------------


  # ------------------ AGREGACIÓN VENTANA DE 1 MES ---------------------------

  # Se filtran los últimos 30 días.
  df = ultimos_n_dias(df_origen, 30, columna_tiempo)
  mask_flag = df['FLAGS']==flag_val
  mask = mask_flag

  #Se dropean los duplicados (teniendo en cuenta la appname y se agrupan por NM).
  agrupados_por_NM = df[mask].drop_duplicates().groupby('NM')
  df_1 = agrupados_por_NM.agg({'FLAGS': 'count', 'NUMFLTS': 'sum'})
  if deb:  print('filtro_1M')

  agrupados_por_NM = df[~mask].groupby('NM')
  f = lambda x: 0
  df_2 = agrupados_por_NM.agg({'FLAGS': f, 'NUMFLTS': f})
  if deb: print('filtro_2')

  df_2.update(df_1)
  df_2 = df_2.rename(columns = {'FLAGS': caract_d + '_1M', 'NUMFLTS': 'NUMFLTS_1M_' + caract_d})
  #if deb: print(df_2)

  if relleno_nans:
    #df_2[caract_d + '_1M'].fillna(df_unificado[caract_d + '_2W'], inplace=True)
    df_2.fillna({caract_d + '_1M': 0, 'NUMFLTS_1M_' + caract_d: 0}, inplace=True)


  df_unificado = pd.merge(df_unificado, df_2, left_index=True, right_index=True, how='left')
  if deb: print('fin 1M')

  # ------------------ FIN AGREGACIÓN VENTANA DE 1 MES ---------------------------


  # ------------------ AGREGACIÓN VENTANA TOTAL INTERVALO ------------------------
  # # Agregado con el total de las medidas
  if deb: print('mask_TOTAL')

  mask_flag = df_origen['FLAGS']==flag_val

  #Se dropean los duplicados (teniendo en cuenta la appname y se agrupan por NM)
  agrupados_por_NM = df_origen[mask_flag].drop_duplicates().groupby('NM')
  df_1 = agrupados_por_NM.agg({'FLAGS': 'count', 'NUMFLTS': 'sum'})

  # Todas las máquinas que tengan una entrada en esta tabla comienzan con un valor en 0
  # para flags y numflts. Esto significa que como al menos en el período total la máquina ha reportado algún
  # flag si luego no reporta el flag en cuestión va a tener un 0 y no un nan.
  # Si luego resulta que reportó uno o más del flag en cuestión este 0 se actualiza por el correcto calculado en
  # el df_1.

  #Se dropean los duplicados (teniendo en cuenta la appname y se agrupan por NM)
  agrupados_por_NM = df_origen.drop_duplicates().groupby('NM')

  f = lambda x: 0
  df_2 = agrupados_por_NM.agg({'FLAGS': f, 'NUMFLTS': f})

  df_2.update(df_1)
  df_2 = df_2.rename(columns = {'FLAGS': caract_d + '_total', 'NUMFLTS': 'NUMFLTS_total_' + caract_d})
  #if deb: print(df_2)

  if relleno_nans:
    df_2.fillna({caract_d + '_total': 0, 'NUMFLTS_total_' + caract_d: 0}, inplace=True)

  # ------------------ FIN AGREGACIÓN VENTANA TOTAL INTERVALO ------------------------


  # Se actualiza el df_unificado con los valores calculados.
  df_unificado = pd.merge(df_unificado, df_2, left_index=True, right_index=True, how='left')

  # Para las máquinas que tengan las 4 características en NaNs se sustituyen los 4 valores por 0.
  mask_nans = df_unificado['FAULTS_1W'].isna() & df_unificado['FAULTS_2W'].isna() & \
       df_unificado['FAULTS_1M'].isna() & df_unificado['FAULTS_total'].isna()

  df_unificado.update(df_unificado[mask_nans][['FAULTS_1W', 'FAULTS_2W', 'FAULTS_1M', 'FAULTS_total']].fillna(0))
```

### Funciones para el balanceo

#### RandomUnderSampler

```python
def usar_RandomUnderSampler():

  global data

  from collections import Counter
  from sklearn.datasets import make_classification
  from imblearn.under_sampling import RandomUnderSampler

  X = data.drop(['target'], axis=1)
  y = data['target']
  X.shape, y.shape

  print('Original dataset shape %s' % Counter(y))

  rus = RandomUnderSampler(random_state=42)
  X_res, y_res = rus.fit_resample(X, y)
  print('Resampled dataset shape %s' % Counter(y_res))

  data = X_res
  data['target'] = y_res

  #data.groupby(['target'])['NM'].count()
```

#### NearMiss

```python
def usar_NearMiss(relleno):

  print("Ejecutando usar_NearMiss()")

  global data

  from collections import Counter
  from imblearn.under_sampling import NearMiss

  if relleno == 'cero':
    # Se reemplazan los NaNs con ceros
    data = data.fillna(0)

  if relleno == 'media':
    # Se reemplazan los NaNs con la media de la columna
    data = data.fillna(data.mean())

  X = data.drop(['target'], axis=1)
  y = data['target']
  X.shape, y.shape

  nm = NearMiss()
  X_res, y_res = nm.fit_resample(X, y)
  print('Resampled dataset shape %s' % Counter(y_res))

  data = X_res
  data['target'] = y_res
```

#### CondensedNearestNeighbour

```python
def usar_CondensedNearestNeighbour():

  print("usar_CondensedNearestNeighbour()")

  global data

  from collections import Counter
  from imblearn.under_sampling import CondensedNearestNeighbour

  X = data.drop(['target'], axis=1)
  y = data['target']
  print('Original dataset shape %s' % Counter(y))

  cnn = CondensedNearestNeighbour(random_state=42)
  X_res, y_res = cnn.fit_resample(X, y)
  print('Resampled dataset shape %s' % Counter(y_res))

  data = X_res
  data['target'] = y_res
```

## Funciones

A continuación se definen una serie de funciones que se reutilizarán para la carga de los datos asociados a las diferentes tablas.

Lista de funciones a definir:

```python

"""
Carga los dataframes de systrack desde los archivos CSV.
Recibe como parámetros la ubicación de los archivos (PATH) y la el formato del nombre de las tablas (tabla).
Se debe ejecutar la funcion para cada tipo de tabla.
Ejemplo RTP_BIOS o RTP_BATTERY
Ejemplo de tabla "*RPT_BIOS*.csv"
"""
def cargar_systrack_csv(PATH, tabla, solo_CXB, d, filtro, numericos, mask='', dtype={}):

  csv_files = glob.glob(os.path.join(PATH, tabla))

  for f in csv_files:
    key = f.split("/")[-1]

    # Si existen columnas de tiempo se importan con el tipo de datos correcto.
    # Las 2 columas de tiempo que aparecen juntas son DOTY y VWTIME
    if 'DOTY' in filtro:
      df_temp = pd.read_csv(f, sep=';', usecols=filtro, parse_dates=['DOTY', 'VWTIME'], dtype=dtype)
    else:
      df_temp = pd.read_csv(f, sep=';', usecols=filtro)
    if not (mask==''):
      #Pendiente a futuro. Cambiar condición por mask
      df_temp = df_temp[(df_temp['WTYPE_ID']==0)]

    # Los campos numericos se pasan a tipo float y cuando encuentra un valor que no puede
    # convertir se pasa a NaN (a veces aparecen valores //N en campos numericos de las tablas)
    df_temp[numericos] =df_temp[numericos].apply(pd.to_numeric, errors='coerce')

    # Float64 da errores en pycaret
    df_temp[numericos] =df_temp[numericos].astype('float32')
    d[key] = proc_systrack(df_temp, solo_CXB=solo_CXB, filtro=filtro)



"""
  proc_systrack()
  Función para procesar los datafrmaes que provienen de systrack
  A partir de un dataframe de Systrack que contiene una colomna SystemId se encuentra el NM (número de máquina).
  Este número es el que nos permitirá vincular al info de Systrack con la de sustituciones de telefónica.
  Ejemplo para el SystemID CXB0920346.LACAIXA.GLC.ES	tenemos que en número de máquina (NM) es 0920346
  Opcionalmente se pueden filtrar las maquinas de Caixa
  """
def proc_systrack(df_systrack, solo_CXB, filtro):

  df_systrack = df_systrack[filtro]

  # Se eliminan las columnas que tienen todos valores null
  df_systrack = df_systrack.dropna(axis=1, how='all')

  # Se eliminan las columnas con valores constantes
  df_systrack = df_systrack.loc[:,df_systrack.nunique() != 1]

  # Se filtran únicamente los SystemId que corresponden ca Caixa
  if solo_CXB:
    df_systrack = df_systrack.dropna(subset = ['SystemId'])
    mask = df_systrack['SystemId'].str.contains("CXB0")
    df_systrack = df_systrack[mask]

  return df_systrack


"""
cargar_etiquetas_csv()
Carga los dataframes de telefonica con el resumen de las sustituciones desde los archivos CSV.

"""
def cargar_etiquetas_csv(PATH, tabla, d):

  csv_files = glob.glob(os.path.join(PATH, tabla))

  for f in csv_files:
    key = f.split("/")[-1]
    d[key] = pd.read_csv(f, sep=';', encoding='latin-1')
```

```python
"""
Función para diagnóstico que devuelve los valores infinitos en un dataframe.
"""
def num_infinitos(df):
  print("Número de inifnitos en el df:\n")
  return df[df == np.inf].count()
```

## Funciones

A continuación se definen una serie de funciones que se reutilizarán para la generación de las características asociadas a las diferentes tablas.

```python
"""
A partir de un diccionario de dataframes del mismo tipo (ejemplo RAW_dfs_RPT_SV_HARDWARE)
genero uno solo unificado.
"""

def unifico_dataframes(df_dict, elimino_duplicados=True):
  primero = True
  for value in df_dict.values():
    print_dimesiones_del_df(value)
    if primero:
        df_unificado = value
        primero = False
    else:
        if elimino_duplicados:
          # Agrego solamente las maquinas que nuevas (o que no estan en el df_unificado)
          value = value[~value.index.isin(df_unificado.index)]
          df_unificado.append(value)
        else:
          df_unificado = df_unificado.append(value)

  print(f"Dimensiones del dataframe unificado:", end=" ")
  print(f"Numero de filas: {len(df_unificado.index)}, Número de columnas {len(df_unificado.columns)}")

  return df_unificado


def map_wguid_to_SystemId(df):

  df_wguid_SystemId = df[['WGUID', 'SystemId']]

  # Algunos dataset cambiaron el formato del WGUI poniendolo entre llaves por lo que se deben eliminar
  df_wguid_SystemId['WGUID'] = df['WGUID'].str.replace("[{}]","")

  # Elimino duplicados
  df_wguid_SystemId = df_wguid_SystemId.drop_duplicates()
  print_dimesiones_del_df(df_wguid_SystemId)

  #df_wguid_SystemId = df_wguid_SystemId.set_index('WGUID')

  return df_wguid_SystemId

# """
# agregar_NM()
# Se agrega el número de máquina NM a un dataframe que no tenga esta columna.
# Se puede hacer usando el SystemId. Si no está presente esta columna se genera a partir del WGUID
# """

def agregar_NM(df, drop_NM_dup=False ,df_map_wguid_to_SystemId=False):

  # Algunos dataset cambiaron el formato del WGUI poniendolo entre llaves por lo que se deben eliminar
  df['WGUID'] = df['WGUID'].str.replace("[{}]","")

  if 'SystemId' not in df:
    df = pd.merge(df, df_map_wguid_to_SystemId, on='WGUID', how='left')

  # Agrego la columna NM obteniendo a partir de la info de SystemID
  df_temp = df.filter(['SystemId'])
  df['NM'] = df_temp['SystemId'].str.extract('(\d+)')

  # Se eliminan las filas que tienen NM nulo. Drop rows that has NaN values on selected columns
  df = df.dropna(subset = ['NM'])

  # La clumna NM se pasa a tipo entero, es importante para eliminar el primer 0 (ya que servicenow no lo tiene)
  df = df.astype({'NM':'uint32'})

  # Opción de elimiar NM duplicados. Evaluar dependiendo del dataset si se debe hacer o no.
  if drop_NM_dup:
    df = df.drop_duplicates(subset='NM')

  # Seteo la columna NM como índice
  df = df.set_index('NM')

  return df

"""
Función para determinar cuál es el dataset más reciente de los que están en un diccionario.
En un diccionario se pueden almacenar varios datasets. La key será el nombre que contiene la fecha
en la que se tomaron los datos del dataset.
Se extrae la fecha del nombre en cada key y se encuentra la mas reciente.
"""
def dataset_mas_reciente(df_dict):
  date_time_str = '00/10/30'
  date_time_obj = datetime.strptime(date_time_str, '%y/%m/%d').date()
  fecha_ultimos_datos = date_time_obj

  for key in df_dict.keys():
    #print(key)
    match = re.search(r'\d{4}_\d{2}_\d{2}', key)
    print(match.group(0))
    date = datetime.strptime(match.group(), '%Y_%m_%d').date()
    if date > fecha_ultimos_datos:
      fecha_ultimos_datos = date
      key_mas_reciente = key

  print(f"Dataset más reciente {key_mas_reciente}")
  return key_mas_reciente
```

```python
"""
ultimos_n_dias: recibe un df y el número de días. El df es una tabla de systrack con el campo DOTY
donde almacena la fecha de las medidas (ej, CPU, MEM, etc) para cada máquina. La función filtra
las medidas de los últimos n días para cada maquina y devuelve ese df filtrado.
"""

def ultimos_n_dias(df, n, columna_tiempo):
  # Se genera ua serie donde para cada NM se obtiene la fecha máxima. (DOTY)
  ult_fecha = df.groupby('NM')[columna_tiempo].transform(max)

  # Se genera un array boleano donde para cada NM de verdadero o falso dependiendo si la fecha DOTY
  # es mayor la la última fecha menos n dias.
  mask_una_semana = df[columna_tiempo] >= ult_fecha - timedelta(days=n)

  # Se devuelve el df filtrado donde se tienen las medidas de los últimos n días para cada NM.
  return df[mask_una_semana]

def agregado_caracteristicas(df_origen, caract, columna_tiempo, relleno_nans = True):

  # Se agrega la característica caract pasada como parámetro al df_unificado, esto utilizando
  # distinas ventanas de tiempo 1 semana (1W), 2 semanas (2W) y 1 mes (1M).
  # Esta versión toma como extremo superior de la ventana de tiempo la última fecha (DOTY) que
  # corresponde al último valor registrado de la máquina.

  # Se agrupa por número de máquina para realizar las agregaciones para cada una de ellas

  deb = False

  global df_unificado

  # mask_1W = (df_origen['DOTY'] > fecha_ini_1W) & (df_origen['DOTY'] <= fecha_fin)
  # mask_2W = (df_origen['DOTY'] > fecha_ini_2W) & (df_origen['DOTY'] <= fecha_fin)
  # mask_1M = (df_origen['DOTY'] > fecha_ini_1M) & (df_origen['DOTY'] <= fecha_fin)


  # ---------------------AGREGACIÓN VENTANA DE 1 SEMANA---------------------------

  # Se filtran los últimos 7 días para cada maquina
  #df = top_n(df_origen, n=7, col='DOTY')

  df = ultimos_n_dias(df_origen, 7, columna_tiempo)


  # Agrupo por NM y ordeno por fecha.

  agrupados_por_NM = df[[columna_tiempo, caract]].groupby('NM')

  df = agrupados_por_NM.min()
  df = df.rename(columns={caract:caract + '_1W_min', columna_tiempo: columna_tiempo +'_1W_' + caract})
  df[caract + '_1W_mean'] = agrupados_por_NM.mean()[caract]
  df[caract + '_1W_std'] = agrupados_por_NM.std()[caract]
  df[caract + '_1W_max'] = agrupados_por_NM.max()[caract]
  df_unificado = pd.merge(df_unificado, df, left_index=True, right_index=True, how='left')

  if relleno_nans:
    if deb: print('relleno 1W')
    print('relleno 1W')
    # df_unificado[caract + '_1W_min'].fillna(0, inplace=True)
    # df_unificado[caract + '_1W_mean'].fillna(0, inplace=True)
    # df_unificado[caract + '_1W_std'].fillna(0, inplace=True)
    # df_unificado[caract + '_1W_max'].fillna(0, inplace=True)
    df_unificado.fillna({caract + '_1W_min': 0, caract + '_1W_mean': 0,
            caract + '_1W_std': 0, caract + '_1W_max': 0}, inplace=True)

  # ------------------ FIN AGREGACIÓN VENTANA DE 1 SEMANA ---------------------------


  # ------------------ AGREGACIÓN VENTANA DE 2 SEMANAS ---------------------------

  # Se filtran los últimos 14 días de cada
  #df = top_n(df_origen, n=14, col='DOTY')
  df = ultimos_n_dias(df_origen, 14, columna_tiempo)


  agrupados_por_NM = df[[columna_tiempo, caract]].groupby('NM')
  df = agrupados_por_NM.min()
  df = df.rename(columns={caract: caract + '_2W_min', columna_tiempo: columna_tiempo + '_2W_' + caract})
  df[caract + '_2W_mean'] = agrupados_por_NM.mean()[caract]
  df[caract + '_2W_std'] = agrupados_por_NM.std()[caract]
  df[caract + '_2W_max'] = agrupados_por_NM.max()[caract]
  df_unificado = pd.merge(df_unificado, df, left_index=True, right_index=True, how='left')

  if relleno_nans:
    if deb: print('relleno 2W')
    print('relleno 2W')
    df_unificado[caract + '_2W_min'].fillna(df_unificado[caract + '_1W_min'], inplace=True)
    df_unificado[caract + '_2W_mean'].fillna(df_unificado[caract + '_1W_mean'], inplace=True)
    df_unificado[caract + '_2W_std'].fillna(df_unificado[caract + '_1W_std'], inplace=True)
    df_unificado[caract + '_2W_max'].fillna(df_unificado[caract + '_1W_max'], inplace=True)

  # ------------------ FIN AGREGACIÓN VENTANA DE 2 SEMANAS ---------------------------


  # --------------------- AGREGACIÓN VENTANA DE 1 MES ---------------------------

  # Se filtran los últimos 30 días de cada
  #df = top_n(df_origen, n=30, col='DOTY')
  df = ultimos_n_dias(df_origen, 30, columna_tiempo)


  # Para poder agregar el último valor medido de la característica, previo a agrupar se ordena por fecha.
  #agrupados_por_NM = df[['DOTY', caract]].sort_values(by='DOTY').groupby('NM')
  agrupados_por_NM = df[[columna_tiempo, caract]].groupby('NM')
  df = agrupados_por_NM.min()
  df = df.rename(columns={caract: caract + '_1M_min', columna_tiempo: columna_tiempo + '_1M_' + caract})
  df[caract + '_1M_mean'] = agrupados_por_NM.mean()[caract]
  df[caract + '_1M_std'] = agrupados_por_NM.std()[caract]
  df[caract + '_1M_max'] = agrupados_por_NM.max()[caract]
  df_unificado = pd.merge(df_unificado, df, left_index=True, right_index=True, how='left')

  if relleno_nans:
    if deb: print('relleno 1M')
    print('relleno 1M')
    df_unificado[caract + '_1M_min'].fillna(df_unificado[caract + '_2W_min'], inplace=True)
    df_unificado[caract + '_1M_mean'].fillna(df_unificado[caract + '_2W_mean'], inplace=True)
    df_unificado[caract + '_1M_std'].fillna(df_unificado[caract + '_2W_std'], inplace=True)
    df_unificado[caract + '_1M_max'].fillna(df_unificado[caract + '_2W_max'], inplace=True)


  # ------------------ FIN AGREGACIÓN VENTANA DE 1 MES ---------------------------



  # Agregado con el total de las medidas
  df = df_origen

  # Para poder agregar el último valor medido de la característica, previo a agrupar se ordena por fecha.
  agrupados_por_NM = df[[columna_tiempo, caract]].sort_values(by=columna_tiempo).groupby('NM')
  df = agrupados_por_NM.min()
  df = df.rename(columns={caract: caract + '_total_min', columna_tiempo: columna_tiempo + '_total_' + caract})
  df[caract + '_total_mean'] = agrupados_por_NM.mean()[caract]
  df[caract + '_total_std'] = agrupados_por_NM.std()[caract]
  df[caract + '_total_max'] = agrupados_por_NM.max()[caract]

  # Se agrega el último valor medido de la característica, este valor se calcula en la ventana mas extensa de tiempo.
  df[caract + '_ult_registro'] = agrupados_por_NM.nth(-1)[caract]
  df[caract + '_fecha_registro_mas_reciente'] = agrupados_por_NM.nth(-1)[columna_tiempo]
  df_unificado = pd.merge(df_unificado, df, left_index=True, right_index=True, how='left')

```

```python
"""
Función para la agregación de las características de la tabla
RPT_VUAPPFAULTSDAILY2 (FAULTS, HANGS y BSODS)
"""
def agregado_caracteristicas_2(df_origen, caract_d, columna_tiempo):

  # Se agrega la característica caract pasda como parámetro al df_unificado, esto utilizando
  # distinas ventanas de tiempo 1 semana (1W), 2 semanas (2W) y 1 mes (1M).
  # Se agrupa por número de máquina para realizar las agregaciones para cada una de ellas

  # Se cuentan los flags y se suman los NUMFLTS

  #deb = False
  deb = True
  relleno_nans = True

  global df_unificado

  if caract_d == 'FAULTS':
    flag_val = 1
  if caract_d == 'HANGS':
    flag_val = 2
  if caract_d == 'BSODS':
    flag_val = 3

  # mask_1W = (df_origen['DOTY'] > fecha_ini_1W) & (df_origen['DOTY'] <= fecha_fin)
  # mask_2W = (df_origen['DOTY'] > fecha_ini_2W) & (df_origen['DOTY'] <= fecha_fin)
  # mask_1M = (df_origen['DOTY'] > fecha_ini_1M) & (df_origen['DOTY'] <= fecha_fin)

  #mask_flag = df_origen['FLAGS']==flag_val

  # ---------------------AGREGACIÓN VENTANA DE 1 SEMANA---------------------------

  # Se filtran los últimos 7 días de cada
  #df = top_n(df_origen, n=7, col='DOTY')
  df = ultimos_n_dias(df_origen, 7, columna_tiempo)


  mask_flag = df['FLAGS']==flag_val
  #mask = mask_1W & mask_flag
  mask = mask_flag

  #agrupados_por_NM = df_origen[mask].groupby('NM')
  #Se dropean los duplicados (teniendo en cuenta la appname y se agrupan por NM)
  agrupados_por_NM = df[mask].drop_duplicates().groupby('NM')
  df_1 = agrupados_por_NM.agg({'FLAGS': 'count', 'NUMFLTS': 'sum'})
  if deb: print('filtro_1')
  #if deb: print(df_1)

  agrupados_por_NM = df[~mask].groupby('NM')
  f = lambda x: 0
  df_2 = agrupados_por_NM.agg({'FLAGS': f, 'NUMFLTS': f})
  if deb: print('filtro_2')
  #if deb: print(df_2)

  df_2.update(df_1)
  df_2 = df_2.rename(columns = {'FLAGS': caract_d + '_1W', 'NUMFLTS': 'NUMFLTS_1W_' + caract_d})

  # if deb: print(df_2)

  if relleno_nans:
      if deb: print('relleno 1W')
      #df_2[caract_d + '_1W'].fillna(0, inplace=True)
      df_2.fillna({caract_d + '_1W': 0, 'NUMFLTS_1W_' + caract_d: 0}, inplace=True)


  df_unificado = pd.merge(df_unificado, df_2, left_index=True, right_index=True, how='left')
  if deb: print('fin 1W')

  # ------------------ FIN AGREGACIÓN VENTANA DE 1 SEMANA ---------------------------


  # ------------------ AGREGACIÓN VENTANA DE 2 SEMANAS ---------------------------

  # Se filtran los últimos 14 días de cada
  df = ultimos_n_dias(df_origen, 14, columna_tiempo)

  mask_flag = df['FLAGS']==flag_val
  mask = mask_flag

  #mask = mask_2W & mask_flag
  #agrupados_por_NM = df_origen[mask].groupby('NM')
  #Se dropean los duplicados (teniendo en cuenta la appname y se agrupan por NM)
  agrupados_por_NM = df[mask].drop_duplicates().groupby('NM')
  df_1 = agrupados_por_NM.agg({'FLAGS': 'count', 'NUMFLTS': 'sum'})
  if deb:  print('filtro_2W_1')

  agrupados_por_NM = df[~mask].groupby('NM')
  f = lambda x: 0
  df_2 = agrupados_por_NM.agg({'FLAGS': f, 'NUMFLTS': f})
  if deb:  print('filtro_2W_2')

  df_2.update(df_1)
  df_2 = df_2.rename(columns = {'FLAGS': caract_d + '_2W', 'NUMFLTS': 'NUMFLTS_2W_' + caract_d})

  if relleno_nans:
    #df_2[caract_d + '_2W'].fillna(df_unificado[caract_d + '_1W'], inplace=True)
    df_2.fillna({caract_d + '_2W': 0, 'NUMFLTS_2W_' + caract_d: 0}, inplace=True)

  # previo al merge relleno todos los NaNs con ceros
  df_2 = df_2.fillna(0)

  df_unificado = pd.merge(df_unificado, df_2, left_index=True, right_index=True, how='left')
  if deb: print('fin 2W')

  # ------------------ FIN AGREGACIÓN VENTANA DE 2 SEMANAS --------------------


  # ------------------ AGREGACIÓN VENTANA DE 1 MES ---------------------------

  # Se filtran los últimos 30 días de cada
  df = ultimos_n_dias(df_origen, 30, columna_tiempo)
  mask_flag = df['FLAGS']==flag_val
  mask = mask_flag

  #mask = mask_1M & mask_flag
  #agrupados_por_NM = df_origen[mask].groupby('NM')
  #Se dropean los duplicados (teniendo en cuenta la appname y se agrupan por NM)
  agrupados_por_NM = df[mask].drop_duplicates().groupby('NM')
  df_1 = agrupados_por_NM.agg({'FLAGS': 'count', 'NUMFLTS': 'sum'})
  if deb:  print('filtro_1M')

  agrupados_por_NM = df[~mask].groupby('NM')
  f = lambda x: 0
  df_2 = agrupados_por_NM.agg({'FLAGS': f, 'NUMFLTS': f})
  if deb: print('filtro_2')

  df_2.update(df_1)
  df_2 = df_2.rename(columns = {'FLAGS': caract_d + '_1M', 'NUMFLTS': 'NUMFLTS_1M_' + caract_d})
  #if deb: print(df_2)

  if relleno_nans:
    #df_2[caract_d + '_1M'].fillna(df_unificado[caract_d + '_2W'], inplace=True)
    df_2.fillna({caract_d + '_1M': 0, 'NUMFLTS_1M_' + caract_d: 0}, inplace=True)


  df_unificado = pd.merge(df_unificado, df_2, left_index=True, right_index=True, how='left')
  if deb: print('fin 1M')

  # ------------------ FIN AGREGACIÓN VENTANA DE 1 MES ---------------------------


  # ------------------ AGREGACIÓN VENTANA TOTAL INTERVALO ------------------------
  # # Agregado con el total de las medidas
  if deb: print('mask_TOTAL')

  mask_flag = df_origen['FLAGS']==flag_val

  #agrupados_por_NM = df_origen[mask_flag].groupby('NM')
  #Se dropean los duplicados (teniendo en cuenta la appname y se agrupan por NM)
  agrupados_por_NM = df_origen[mask_flag].drop_duplicates().groupby('NM')
  df_1 = agrupados_por_NM.agg({'FLAGS': 'count', 'NUMFLTS': 'sum'})

  # Todas las máquinas que tengan una entrada en esta tabla comienzan con un valor en 0
  # para flags y numflts. Esto significa que como al menos en el período total la máquina ha reportado algún
  # flag si luego no reporta el flag en cuestión va a tener un 0 y no un nan.
  # Si luego resulta que reportó uno o más del flag en cuestión este 0 se actualiza por el correcto calculado en
  # el df_1.
  #agrupados_por_NM = df_origen.groupby('NM')
  #Se dropean los duplicados (teniendo en cuenta la appname y se agrupan por NM)
  agrupados_por_NM = df_origen.drop_duplicates().groupby('NM')

  f = lambda x: 0
  df_2 = agrupados_por_NM.agg({'FLAGS': f, 'NUMFLTS': f})

  df_2.update(df_1)
  df_2 = df_2.rename(columns = {'FLAGS': caract_d + '_total', 'NUMFLTS': 'NUMFLTS_total_' + caract_d})
  #if deb: print(df_2)

  if relleno_nans:
    #df_2[caract_d + '_total'].fillna(df_unificado[caract_d + '_1M'], inplace=True)
    df_2.fillna({caract_d + '_total': 0, 'NUMFLTS_total_' + caract_d: 0}, inplace=True)

  # ------------------ FIN AGREGACIÓN VENTANA TOTAL INTERVALO ------------------------


  # Se actualiza el df_unificado con los valores calculados.
  df_unificado = pd.merge(df_unificado, df_2, left_index=True, right_index=True, how='left')


  # Para las máquinas que tengan las 4 características en NaNs se sustituyen los 4 valores por 0.
  mask_nans = df_unificado['FAULTS_1W'].isna() & df_unificado['FAULTS_2W'].isna() & \
       df_unificado['FAULTS_1M'].isna() & df_unificado['FAULTS_total'].isna()

  df_unificado.update(df_unificado[mask_nans][['FAULTS_1W', 'FAULTS_2W', 'FAULTS_1M', 'FAULTS_total']].fillna(0))
```

## Agregado de características

En esta instancia se recorren las distintas tablas y se procesan sus campos para generar y agregar las características correspondientes al dataset unificado.

Para el agregado de características se utilizan tablas con un índice común que será el número de máquina ('NM'). Se dispone de la función agregar_NM para agregar el número de máquina como índice en caso de no existir en la tabla.

1. Cuando provienenen de datasets que tienen una entrada por máquina (ejemplo RTP_BATTERY).
